```python
# Step 1: Define base seasons
base_seasons = {"зима", "весна", "лето", "осень", "межсезонье"}

# Step 2: Create rename mapping based on 2024 / 2025
rename_mapping = {}
for i, row in filtered_nodes[filtered_nodes["label"] == "Сезон"].iterrows():
    name = row["name"]
    if "2024" in name:
        rename_mapping[row["id"]] = "Тренд:2024"
    elif "2025" in name:
        rename_mapping[row["id"]] = "Тренд:2025"

# Step 3: Apply mapping to filtered_nodes
filtered_nodes["id"] = filtered_nodes["id"].replace(rename_mapping)
filtered_nodes["name"] = filtered_nodes["name"].replace(rename_mapping)
filtered_nodes.loc[filtered_nodes["id"].isin(rename_mapping.values()), "label"] = "Тренд"

# Step 4: Determine removed_seasons
# Get all Сезон nodes again *after* rename
season_nodes_post = filtered_nodes[filtered_nodes["label"] == "Сезон"]

removed_seasons = season_nodes_post[
    ~season_nodes_post["name"].isin(base_seasons)
]["id"].unique().tolist()

# Step 5: Drop removed_seasons and deduplicate
filtered_nodes_after_season_nodes_removal = (
    filtered_nodes[~filtered_nodes["id"].isin(removed_seasons)]
    .drop_duplicates(subset=["id", "name", "label"])
    .reset_index(drop=True)
)

# Step 6: Final trends list
new_trends = sorted(set(rename_mapping.values()))
```

```python
# Step 1: Work only with trend nodes
trend_mask = filtered_nodes_after_season_nodes_removal["label"] == "Тренд"
trend_nodes = filtered_nodes_after_season_nodes_removal[trend_mask].copy()

# Step 2: Build reverse mapping and rename 2024/2025 trends only
reverse_mapping_trends = {}

def trend_standardizer(row):
    original_id = row["id"]
    if "2024" in row["name"]:
        row["id"] = "Тренд:2024"
        row["name"] = "2024"
        row["label"] = "Тренд"
        reverse_mapping_trends[original_id] = "Тренд:2024"
        return row
    elif "2025" in row["name"]:
        row["id"] = "Тренд:2025"
        row["name"] = "2025"
        row["label"] = "Тренд"
        reverse_mapping_trends[original_id] = "Тренд:2025"
        return row
    else:
        return None  # mark for removal

# Apply and remove non-matching
trend_nodes_cleaned = trend_nodes.apply(trend_standardizer, axis=1).dropna()

# Step 3: Keep non-trend nodes untouched
non_trend_nodes = filtered_nodes_after_season_nodes_removal[~trend_mask].copy()

# Step 4: Concatenate back
filtered_nodes_after_season_trend_nodes_removal = pd.concat(
    [non_trend_nodes, trend_nodes_cleaned],
    ignore_index=True
).drop_duplicates(subset=["id", "name", "label"]).reset_index(drop=True)

# Step 5: Final values
new_trends = ["Тренд:2024", "Тренд:2025"]
```
